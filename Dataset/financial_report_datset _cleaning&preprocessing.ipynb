```python
import pandas as pd
import numpy as np
```

```python
df2 = pd.read_csv('IndianFinancialNews.csv')
```

```python
df2.head()  
```

```python
#convert the date column to datetime format
df2['Date'] = pd.to_datetime(df2['Date'])
```

```python
#drop Unnamed: 0 column
df2.drop(['Unnamed: 0'], axis=1, inplace=True)
df2.head()
```

```python
#Create 2 columns category and main_category with all values as financial documents
df2['category'] = 'Financial documents'
df2['main_category'] = 'Financial documents'
df2.head()
```

```python
#reaarange the columns in the dataframe
df2 = df2[['Title', 'main_category', 'category', 'Date', 'Description']]
```

```python
#rename the columns
df2.rename(columns={'Date':'date', 'Description':'description'}, inplace=True)
```

```python
import nltk
nltk.download('stopwords')

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string

def preprocess_text(text):
    if isinstance(text, str):
        lemmatizer = WordNetLemmatizer()
        stop_words = set(stopwords.words("english"))
        punctuations = set(string.punctuation)

        # Your preprocessing steps
        text = text.lower()
        tokens = word_tokenize(text)
        tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words and token not in punctuations and token.isalpha()]

        return ' '.join(tokens)
    else:
        # Handle non-string values, you can return an empty string or handle it in another way based on your requirements.
        return ''

# Apply the preprocessing function to the 'Title' and 'description' columns
df2['Title_clean'] = df2['Title'].apply(preprocess_text)
df2['description_clean'] = df2['description'].apply(preprocess_text)

# Save the cleaned dataset to a CSV file
df2.to_csv('financial_clean_dataset.csv', index=False)
```

```python
df3 = pd.read_csv('financial_clean_dataset.csv')
df3.head()
```

```python
#drop Title and description columns
df3.drop(['Title', 'description'], axis=1, inplace=True)
#rename the columns
df3.rename(columns={'Title_clean':'Title', 'description_clean':'description'}, inplace=True)
#reoorder the columns   
df3 = df3[['Title', 'main_category', 'category', 'date', 'description']]
```
