```python
import pandas as pd
import numpy as np

data = pd.read_csv('../Dataset/combined_final_dataset.csv')
data.head(10)
```

```python
#Finding keywords from the description column
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

# Fit and transform the 'description' column
X = vectorizer.fit_transform(data['description'])

# Find the highest frequency words in each row
max_freq_words = []
for i, description in enumerate(data['description']):
    row = X[i].toarray().flatten()  # Convert to an array and flatten it
    sorted_word_indices = row.argsort()[-5:][::-1]  # Get indices of top 5 words
    top_5_words = [list(vectorizer.vocabulary_.keys())[list(vectorizer.vocabulary_.values()).index(idx)] for idx in sorted_word_indices]
    max_freq_words.append(top_5_words)

# Add a new 'Keyword' column to the DataFrame with the 5 maximum frequency words
data['Keyword'] = max_freq_words
```

```python
df = pd.read_csv('../Dataset/combined_final_dataset.csv')
df.head(10)
```

```python
import yake

kw_extractor = yake.KeywordExtractor()

deduplication_threshold = 0.3
deduplication_algo = 'seqm'
windowSize = 1
numOfKeywords = 5

kw_extractor = yake.KeywordExtractor(dedupLim=deduplication_threshold, dedupFunc=deduplication_algo, windowsSize=windowSize, top=numOfKeywords)

# Extract keywords from the 'description' column and store them in a new 'Keyword' column
df['Keyword'] = df['description'].apply(lambda x: [kw[0] for kw in kw_extractor.extract_keywords(x)])

df.head()
```
